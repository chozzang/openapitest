```python
import pandas as pd
import folium

# Excel 파일 불러오기
filePath = r'지역별 폭염일수.xlsx'
df_from_excel = pd.read_excel(filePath, engine='openpyxl', header=4)  # 5번째 행을 헤더로 지정

# 열 이름 확인 및 재지정(필요 시)
df_from_excel.columns = ['이름', '경도', '위도', '폭염일수']

# 위도와 경도를 숫자로 변환
df_from_excel['경도'] = pd.to_numeric(df_from_excel['경도'], errors='coerce')
df_from_excel['위도'] = pd.to_numeric(df_from_excel['위도'], errors='coerce')

# 지도 생성
map = folium.Map(location=[37, 127], zoom_start=7)

# 마커 추가
for i in range(len(df_from_excel)):
    name = df_from_excel.loc[i, '이름']  # 지역 이름
    longitude = df_from_excel.loc[i, '경도']  # 경도
    latitude = df_from_excel.loc[i, '위도']  # 위도
    heat_days = df_from_excel.loc[i, '폭염일수']  # 폭염일수

    # 값이 유효한 경우에만 마커 추가
    if pd.notnull(latitude) and pd.notnull(longitude):
        folium.Marker(
            location=[latitude, longitude],
            popup=f"{name} ({heat_days}일)",
            icon=folium.Icon(color='blue')
        ).add_to(map)

# HTML 파일로 저장
map.save('폭염지도.html')
```
